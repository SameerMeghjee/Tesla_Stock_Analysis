```python
#!pip install yfinance

import yfinance as yf

# Create a Ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Get historical market data for maximum period
tesla_data = tesla.history(period="max")

# Reset index and display the first 5 rows
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Load saved HTML file
with open("tesla_revenue.html", "r", encoding="utf-8") as f:
    html_data = f.read()

soup = BeautifulSoup(html_data, "html.parser")

# Find Tesla Quarterly Revenue table
for table in soup.find_all("table"):
    if "Tesla Quarterly Revenue" in table.text:
        tesla_revenue = pd.read_html(str(table))[0]
        break

# Clean data
tesla_revenue.columns = ["Date", "Revenue"]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(r"\$", "", regex=True)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(",", "", regex=False)
tesla_revenue = tesla_revenue.dropna()
tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].astype(float)

tesla_revenue.head()
```

```python
import plotly.graph_objects as go

def make_graph(stock_data, revenue_data, stock_name):
    fig = go.Figure()
    
    fig.add_trace(go.Scatter(
        x=stock_data['Date'],
        y=stock_data['Close'],
        name="Stock Price"
    ))
    
    fig.add_trace(go.Bar(
        x=revenue_data['Date'],
        y=revenue_data['Revenue'],
        name="Revenue"
    ))
    
    fig.update_layout(
        title=stock_name + " Stock Price & Revenue",
        xaxis_title="Date",
        yaxis_title="Price / Revenue (USD Millions)",
        barmode="group"
    )
    
    fig.show()

# Plot Tesla
make_graph(tesla_data, tesla_revenue, "Tesla")
```

```python
# Plot GameStop
make_graph(gme_data, gme_revenue, "GameStop")
```
